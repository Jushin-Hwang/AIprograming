# 필기체 숫자 이미지를 분류해보자

# 1. 데이터 세트 읽어들이기

```python
import matplotlib.pyplot as plt
from sklearn import datasets, metrics
from sklearn.model_selection import train_test_split

digits = datasets.load_digits()
plt.imshow(digits.images[0], cmap = plt.cm.gray_r, interpolation = 'nearest')
```

# 2. 데이터 평탄화 : 2차원 배열을 1차원으로 변경

```python
n_samples = len(digits.images)
data = digits.images.reshape((n_samples, -1))
```

# 3. 훈련 데이터와 테스트 데이터

```python
# split X_train, Y_train, X_test, Y_test

X_train, X_test, y_train, y_test = train_test_split(data, digits.target, test_size = 0.2)
```

# 4_1. KNN-Algorithm 모델

```python
from sklearn.neighbors import KNeighborsClassifier

clf = KNeighborsClassifier(n_neighbors=6)

# 학습
clf.fit(X_train, y_train)

# 테스트 데이터로 예측
y_pred= clf.predict(X_test)

# 정확도 계산
scores = metrics.accuracy_score(y_test, y_pred)
print(scores)

# 이미지를 출력하고 무엇으로 예측하였는지 출력
plt.imshow(X_test[10].reshape(8, 8), cmap = plt.cm.gray_r, interpolation = 'nearest')
y_pred = clf.predict([X_test[10]])
print(y_pred)
```

# 4_2. Logistic Regression 모델

```python
from sklearn.linear_model import LogisticRegression
import numpy as np

clf = LogisticRegression()

# 학습
clf.fit(X_train, y_train)

# 테스트 데이터로 예측
y_pred= clf.predict(X_test)

# 정확도 계산
scores = metrics.accuracy_score(y_test, y_pred)
print(scores)

# 이미지를 출력하고 무엇으로 예측하였는지 출력
plt.imshow(X_test[10].reshape(8, 8), cmap = plt.cm.gray_r, interpolation = 'nearest')
y_pred = clf.predict([X_test[10]])
```

# 4_3. decision tree 모델

```python
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier(random_state=0)

# 학습
clf.fit(X_train, y_train)

# 테스트 데이터로 예측
y_pred= clf.predict(X_test)

# 정확도 계산
scores = metrics.accuracy_score(y_test, y_pred)
print(scores)

# 이미지를 출력하고 무엇으로 예측하였는지 출력
plt.imshow(X_test[10].reshape(8, 8), cmap = plt.cm.gray_r, interpolation = 'nearest')
y_pred = clf.predict([X_test[10]])
print(y_pred)
```

# 4_4. Random Forest 모델

```python
# random forest 인공지능 모델을 생성하고 모델을 훈련, 예측, 결과값 출력

# RandomForestClassifier : 데이터셋에 사용할 Random Forest 모델을 생성하는 함수
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(random_state=0)

# 학습
clf.fit(X_train, y_train)

# 테스트 데이터로 예측
y_pred= clf.predict(X_test)

# 정확도 계산
scores = metrics.accuracy_score(y_test, y_pred)
print(scores)

# 이미지를 출력하고 무엇으로 예측하였는지 출력
plt.imshow(X_test[10].reshape(8, 8), cmap = plt.cm.gray_r, interpolation = 'nearest')
y_pred = clf.predict([X_test[10]])
print(y_pred)
```

# 4_5. Neural Network 모델

```python
from sklearn.neural_network import MLPClassifier

clf = MLPClassifier(random_state=0)

# 학습
clf.fit(X_train, y_train)

# 테스트 데이터로 예측
y_pred= clf.predict(X_test)

# 정확도 계산
scores = metrics.accuracy_score(y_test, y_pred)
print(scores)

# 이미지를 출력하고 무엇으로 예측하였는지 출력
plt.imshow(X_test[10].reshape(8, 8), cmap = plt.cm.gray_r, interpolation = 'nearest')
y_pred = clf.predict([X_test[10]])
print(y_pred)
```
