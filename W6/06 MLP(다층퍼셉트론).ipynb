# 1. 활성화함수 구현

```python
# 계단함수(Step function)

import numpy as np

def step(x) :
    if x > 0.000001 :
        return 1
    else :
        return 0
    
def step(x) :
    result = x > 0.000001
    return result.astype(int)

# 그래프로 확인

import matplotlib.pyplot as plt

x = np.arange(-10.0, 10.0, 0.1)
y = step(x)
plt.plot(x, y)
plt.show()
```

```python
# 시그모이드 함수(Sigmoid function)

def sigmoid(x) :
    return 1.0 / (1.0 + np.exp(-x))

# 그래프로 확인

import matplotlib.pyplot as plt

x = np.arange(-10.0, 10.0, 0.1)
y = sigmoid(x)
plt.plot(x, y)
plt.show()
```

```python
# tanh 함수

x = np.linspace(-np.pi, np.pi, 60)
y = np.tanh(x)
plt.plot(x, y)
plt.show()
```

```python
# ReLU 함수

def relu(x) :
    return np.maximum(x, 0)

# 그래프로 확인

import matplotlib.pyplot as plt

x = np.arange(-10.0, 10.0, 0.1)
y = relu(x)
plt.plot(x, y)
plt.show()
```

# 2. MLP 순방향 패스

```python
import numpy as np

# 시그모이드 함수 구현
def sigmoid(x) :
    return 1.0 / (1.0 + np.exp(-x))

# 시그모이드 함수의 미분치
def sigmoid_deriv(x) :
    return x * (1-x)

# 입력 유닛의 개수, 은닉 유닛의 개수, 출력 유닛의 개수
inputs, hiddens, outputs = 2, 2, 1
learning_rate = 0.2

# 훈련 샘플과 정답
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
T = np.array([[0], [1], [1], [0]])

W1 = np.array([[0.10, 0.20],
              [0.30, 0.40]])
W2 = np.array([[0.50], [0.60]])
B1 = np.array([0.1, 0.2])
B2 = np.array([0.3])

# 순방향 전파 계산
def predict(x) :
    layer0 = x
    Z1 = np.dot(layer0, W1) + B1
    layer1 = sigmoid(Z1)
    Z2 = np.dot(layer1, W2) + B2
    layer2 = sigmoid(Z2)
    return layer0, layer1, layer2

def test() :
    for x, y in zip(X, T) :
        x = np.reshape(x, (1, -1))
        layer0, layer1, layer2 = predict(x)
        print(x, y, layer2)

test()
```

# 5. Numpy로 MLP 구현

```python
# 순방향 전파

# Sigmoid function
def sigmoid(x) :
    return 1.0 / (1.0 + np.exp(-x))

# Sigmoid function 미분
def sigmoid_deriv(x) :
    return x*(1-x)

# 입력유닛의 개수, 은닉유닛의 개수, 출력유닛의 개수
inputs, hiddens, outputs = 2, 2, 1
learning_rate = 0.2

# 훈련 샘플과 정답
X = np.array([[0,0], [0,1], [1,0], [1,1]])
T = np.array([[1], [0], [0], [1]])

W1 = np.array([[0.10, 0.20], [0.30, 0.40]])
W2 = np.array([[0.50], [0.60]])
B1 = np.array([0.1, 0.2])
B2 = np.array([0.3])

# 순방향 전파 계산
def predict(x) :
    layer0 = x
    Z1 = np.dot(layer0, W1) + B1
    layer1 = sigmoid(Z1)
    Z2 = np.dot(layer1, W2) + B2
    layer2 = sigmoid(Z2)
    return layer0, layer1, layer2
```

```python
# 역방향 전파 계산
def fit() :
    global W1, W2, B1, B2
    for i in range(90000) :
        for x, y in zip(X, T) :
            x = np.reshape(x, (1, -1))
            y = np.reshape(y, (1, -1))

            layer0, layer1, layer2 = predict(x)
            layer2_error = layer2 - y
            layer2_delta = layer2_error * sigmoid_deriv(layer2)
            layer1_error = np.dot(layer2_delta, W2.T)
            layer1_delta = layer1_error * sigmoid_deriv(layer1)

            W2 += -learning_rate * np.dot(layer1.T, layer2_delta)
            W1 += -learning_rate * np.dot(layer0.T, layer1_delta)
            B2 += -learning_rate * np.sum(layer2_delta, axis = 0)
            B1 += -learning_rate * np.sum(layer1_delta, axis = 0)
```

```python
# 테스트
def test() :
    for x, y in zip(X, T) :
        x = np.reshape(x, (1, -1))
        layer0, layer1, layer2 = predict(x)
        print(x, y, layer2)
```

```python
# 학습, 테스트
fit()
test()
```
