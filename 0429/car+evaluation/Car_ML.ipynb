```python
# import module

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.linear_model import LogisticRegression
```

```python
# read file

file_path = r'C:\Users\황주신\Documents\University\3학년\1학기\01. 인공지능개론\03. 실습, 과제\W9\car+evaluation\car.data'

index = ['Buying', 'Maint', 'Doors', 'Persons', 'Lug_boot', 'Safety', 'Class']

data = pd.read_csv(file_path, header = None)
df = pd.DataFrame(data)

df.columns = index

df
```

```python
# 결측치 확인
df.isnull().sum()
```

```python
# module import

from sklearn.preprocessing import LabelEncoder

# 라벨 인코딩 함수
def label_encode(data, column):
    label_encoder = LabelEncoder()
    data[column] = label_encoder.fit_transform(data[column])
    return data

# 데이터(data)와 인코딩할 열(column)을 전달하여 label encoding 수행
data = label_encode(data, 0)
data = label_encode(data, 1)
data = label_encode(data, 4)
data = label_encode(data, 5)
# data = label_encode(data, '')

print(data)

# 인코딩된 열을 DataFrame에 추가
df['Buying_encoded'] = data[0]
df['Maint_encoded'] = data[1]
df['Lug_boot_encoded'] = data[4]
df['Safety_encoded'] = data[5]

# 인코딩 되지 않은 열을 DataFrame에서 삭제
df_encoded = df.drop('Buying', axis = 1)
df_encoded = df_encoded.drop('Maint', axis = 1)
df_encoded = df_encoded.drop('Lug_boot', axis = 1)
df_encoded = df_encoded.drop('Safety', axis = 1)

df_encoded
```

```python
# doors에 5more을 5로 변경하기
df_encoded['Doors'] = df_encoded['Doors'].replace('5more', 5)

# persons에 more을 6으로 변경하기
df_encoded['Persons'] = df_encoded['Persons'].replace('more', 6)
```

```python
X = df_encoded.drop('Class', axis = 1)
Y = df['Class']

print(X.head())
print()
print(Y.value_counts())
```

```python
# split X_train, Y_train, X_test, Y_test

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.linear_model import LogisticRegression
import numpy as np

clf = LogisticRegression()

# 학습
clf.fit(X_train, Y_train)

# 테스트 데이터로 예측
y_pred= clf.predict(X_test)

# 정확도 계산
print ("\n--- Logistic Regression Classifier ---")
print (accuracy_score(Y_test, y_pred))
print (confusion_matrix(Y_test, y_pred))
```

```python
# DecisionTree

from sklearn.tree import DecisionTreeClassifier
import numpy as np

clf = DecisionTreeClassifier(random_state=0)

# 학습
clf.fit(X_train, Y_train)

# 테스트 데이터로 예측
y_pred= clf.predict(X_test)

# 정확도 계산
print ("\n--- Logistic Regression Classifier ---")
print (accuracy_score(Y_test, y_pred))
print (confusion_matrix(Y_test, y_pred))
```
