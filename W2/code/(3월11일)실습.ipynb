# CSV 파일 불러오기

```python
# data processing을 위한 모듈 import

# pandas : 데이터 조작 및 분석을 위한 소프트웨어 라이브러리
import pandas as pd
```

```python
# csv 파일 불러오기

data1 = pd.read_csv("../data/breast_cancer.csv", index_col=0) # 유방암 data
data2 = pd.read_csv("../data/diabetes.csv", index_col = 0) # 당뇨 data

df1 = data1
df2 = data2
```

# 러신머닝_준비

```python
# 러신머닝을 위한 모듈 import

# sklearn : 파이썬 프로그래밍 언어용 자유 소프트웨어 기계 학습 라이브러리
# train_test_split(X, Y, test_size = , random_state = ) :
    # X, Y => 입력변수 , test_size => 트레이닝과 테스트 데이터의 비율, random_state => 랜덤한 선택으로 데이터 값을 분류
from sklearn.model_selection import train_test_split
# sklearn.metrics : scikit-learn의 분류 성능 평가
from sklearn.metrics import accuracy_score, confusion_matrix
```

# 러신머닝_breast_cancer

```python
# breast_cancer data 확인

df1
```

```python
# label 제거해서 X에 저장

X=df1.drop('label',axis=1)
X.head()
```

```python
# label Y에 저장

Y=df1['label']
Y.value_counts()
```

```python
# X, Y를 8:2로 나눠 training data와 test data로 나눔

X_train, X_test, Y_train, Y_test = train_test_split(X, Y,test_size=0.2,random_state=0)
```

```python
# logistic regression 인공지능 모델을 생성하고 모델을 훈련, 예측, 결과값 출력

# LogisticRegression : 데이터셋에 사용할 Logistic Regression 모델을 생성하는 함수
from sklearn.linear_model import LogisticRegression

clf_lr = LogisticRegression(random_state=0)

# .fit() : 훈련 함수
clf_lr.fit(X_train, Y_train)

# .predict() : 예측 함수
pred_lr = clf_lr.predict(X_test)

# 실제 결과와 예측 결과를 비교
print ("\n--- Logistic Regression Classifier ---")
print (accuracy_score(Y_test, pred_lr))
print (confusion_matrix(Y_test, pred_lr))
```

# 러신머닝_diabetes

```python
# diabetes data 확인

df2
```

```python
# Outcome 제거해서 X에 저장

X = df2.drop('Outcome',axis = 1)
X.head()
```

```python
# Outcome Y에 저장

Y = df2['Outcome']
Y.value_counts()
```

```python
# X, Y를 8:2로 나눠 training data와 test data로 나눔

X_train, X_test, Y_train, Y_test = train_test_split(X, Y,test_size=0.2,random_state=0)
```

```python
# logistic regression 인공지능 모델을 생성하고 모델을 훈련, 예측, 결과값 출력

# LogisticRegression : 데이터셋에 사용할 Logistic Regression 모델을 생성하는 함수
from sklearn.linear_model import LogisticRegression

clf_lr = LogisticRegression(random_state=0)
clf_lr.fit(X_train, Y_train) # .fit() : 훈련 함수

pred_lr = clf_lr.predict(X_test) # .predict() : 예측 함수

# 실제 결과와 예측 결과를 비교
print ("\n--- Logistic Regression Classifier ---")
print (accuracy_score(Y_test, pred_lr))
print (confusion_matrix(Y_test, pred_lr))
```

```python
# neural network 인공지능 모델을 생성하고 모델을 훈련, 예측, 결과값 출력

# MSPClassfier : 데이터셋에 사용할 Neural Network 모델을 생성하는 함수
from sklearn.neural_network import MLPClassifier

clf_nn = MLPClassifier(random_state=0)
clf_nn.fit(X_train, Y_train)

pred_nn = clf_nn.predict(X_test)

print ("\n--- Neural Network Classifier ---")
print (accuracy_score(Y_test, pred_nn))
print (confusion_matrix(Y_test, pred_nn))
```

```python
# decision tree 인공지능 모델을 생성하고 모델을 훈련, 예측, 결과값 출력

# DecisionTreeClassifier : 데이터셋에 사용할 Decision Tree 모델을 생성하는 함수
from sklearn.tree import DecisionTreeClassifier

clf_dt = DecisionTreeClassifier(random_state=0)
clf_dt.fit(X_train, Y_train)

pred_dt = clf_dt.predict(X_test)

print ("\n--- Decision Tree Classifier ---")
print (accuracy_score(Y_test, pred_dt))
print (confusion_matrix(Y_test, pred_dt))
```

```python
# random forest 인공지능 모델을 생성하고 모델을 훈련, 예측, 결과값 출력

# RandomForestClassifier : 데이터셋에 사용할 Random Forest 모델을 생성하는 함수
from sklearn.ensemble import RandomForestClassifier


rf_clf = RandomForestClassifier(random_state=0)
rf_clf.fit(X_train, Y_train)

pred = rf_clf.predict(X_test)

print ("\n--- Radom Forest ---")
print(accuracy_score(Y_test,pred))
print (confusion_matrix(Y_test, pred))
```
