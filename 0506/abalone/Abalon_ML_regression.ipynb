# 1) 데이터 읽기

```python
# read file
import pandas as pd

file_path = r'C:\Users\황주신\Documents\University\3학년\1학기\01. 인공지능개론\03. 실습, 과제\W10\abalone\abalone.data'

index = ['Sex', 'Length', 'Diameter', 'Height', 'Whole weight', 'Shucked weight', 'Viscera weight', 'Shell weight', 'Rings']

data = pd.read_csv(file_path, header = None)
df = pd.DataFrame(data)

df.columns = index

df
```

```python
# module import

from sklearn.preprocessing import LabelEncoder

# 라벨 인코딩 함수
def label_encode(data, column):
    label_encoder = LabelEncoder()
    data[column] = label_encoder.fit_transform(data[column])
    return data

# 데이터(data)와 인코딩할 열(column)을 전달하여 label encoding 수행
data = label_encode(data, 0)
# data = label_encode(data, '')

print(data)

# 인코딩된 열을 DataFrame에 추가
df['Sex(encoded)'] = data[0]

# 인코딩 되지 않은 열을 DataFrame에서 삭제
df = df.drop('Sex', axis = 1)

df
```

# 5) X, Y 나누기

```python
X = df.drop('Rings', axis = 1)
Y = df['Rings']

print(X.head())
print()
print(Y.value_counts())
```

# 6) 데이터 스케일링

```python
from sklearn.preprocessing import StandardScaler
# from sklearn.preprocessing import MinMaxScaler


scaler = StandardScaler()
X_scaled_data = scaler.fit_transform(X)

X = pd.DataFrame(X_scaled_data,
                            index = X.index,
                            columns = X.columns) 

X
```

# 7) Train데이터, Test데이터 나누기

```python
# split X_train, Y_train, X_test, Y_test
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

# 8) 모델 정의, 학습, 예측

```python
from sklearn.linear_model import LinearRegression
import numpy as np
from sklearn.metrics import mean_squared_error
from sklearn.metrics import accuracy_score, confusion_matrix

clf = LinearRegression()

# 학습
clf.fit(X_train, Y_train)

# 테스트 데이터로 예측
y_pred= clf.predict(X_test)

# 실수형으로 예측된 y_pred 값을 가까이 있는 int형으로 수정
y_pred = [round(prob) for prob in y_pred]

# 정확도 계산
print ("\n--- Logistic Regression Regressor ---")
mse = np.sqrt(mean_squared_error(y_pred, Y_test))
print('평균제곱근오차', mse)
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
from sklearn.metrics import mean_squared_error
from sklearn.metrics import accuracy_score, confusion_matrix

clf = LinearRegression()

# 학습
clf.fit(X_train, Y_train)

# 테스트 데이터로 예측
y_pred= clf.predict(X_test)

# 정확도 계산
print ("\n--- Logistic Regression Regressor ---")
mse = np.sqrt(mean_squared_error(y_pred, Y_test))
print('평균제곱근오차', mse)
```

```python
# DecisionTree

from sklearn.tree import DecisionTreeRegressor
import numpy as np

clf = DecisionTreeRegressor()

# 학습
clf.fit(X_train, Y_train)

# 테스트 데이터로 예측
y_pred= clf.predict(X_test)

# 정확도 계산
print ("\n--- Decision Tree Regressor ---")
mse = np.sqrt(mean_squared_error(y_pred, Y_test))
print('평균제곱근오차', mse)
```

```python
# Random Forest

from sklearn.ensemble import RandomForestRegressor
import numpy as np

clf = RandomForestRegressor()

# 학습
clf.fit(X_train, Y_train)

# 테스트 데이터로 예측
y_pred= clf.predict(X_test)

# 정확도 계산
print ("\n--- Random Forest Regressor ---")
mse = np.sqrt(mean_squared_error(y_pred, Y_test))
print('평균제곱근오차', mse)
```

```python
# KNN Algorithm (KNN elbow 구하기)

from sklearn.neighbors import KNeighborsRegressor

# 다양한 이웃 수(K)에 대해 정확도를 저장할 리스트 초기화
neighbors = list(range(1, 21))
accuracies = []

# 다양한 이웃 수에 대해 모델을 훈련하고 정확도를 평가
for k in neighbors:
    knn = KNeighborsRegressor(n_neighbors=k)
    knn.fit(X_train, Y_train)
    score = knn.score(X_test, Y_test)
    accuracies.append(score)

# 정확도를 시각화
plt.plot(neighbors, accuracies, marker='o')
plt.xlabel('K')
plt.ylabel('Accuracy')
plt.title('KNN Elbow Method')
plt.xticks(range(0, 21, 5))
plt.show()
```

```python
# KNN Algorithm

model = KNeighborsRegressor(n_neighbors = 7)
model.fit(X_train, Y_train)

ky_pred = model.predict(X_test)
mse = np.sqrt(mean_squared_error(ky_pred, Y_test))
print('평균제곱근오차', mse)
```
